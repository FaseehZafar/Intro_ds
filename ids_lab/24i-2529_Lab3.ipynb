```python
# question 4
# Task 4 (Simplified)
def classify_emails(emails):
    categories = {
        "Important": ['urgent', 'asap', 'important', 'meeting', 'project', 'deadline', 'action required'],
        "Promotions": ['sale', 'discount', 'offer', 'deal', 'promotion', 'buy now', 'limited time'],
        "Spam": ['lottery', 'prize', 'winner', 'free', 'click here', 'subscribe', 'risk free']
    }

    classified = {"Important": [], "Promotions": [], "Spam": []}

    for email in emails:
        lower_email = email.lower()
        placed = False
        for category, keywords in categories.items():
            if any(word in lower_email for word in keywords):
                classified[category].append(email)
                placed = True
                break
        if not placed:
            classified["Important"].append(email)  # default

    # Printing result
    for category, mails in classified.items():
        print(f"\n{category} Emails:")
        for m in mails:
            print("-", m)

    return classified


# Example usage
emails = [
    "Urgent: Project deadline is tomorrow",
    "Limited time offer! Buy now and save 50%",
    "Congratulations! You are a lottery winner",
    "Meeting rescheduled to next week",
    "Exclusive discount for subscribers",
    "Action required: update your account details"
]

classify_emails(emails)
```

```python
# question 5

# Task #05 - Simple Version
inventory = []

def add_book(title, author, quantity, price):
    book = {"title": title, "author": author, "quantity": quantity, "price": price}
    inventory.append(book)

def search_book(title):
    for book in inventory:
        if book["title"].lower() == title.lower():
            return book
    return None

def update_quantity(title, new_quantity):
    book = search_book(title)
    if book:
        book["quantity"] = new_quantity
        print("Updated", title, "quantity to", new_quantity)
    else:
        print("Book not found!")

def calculate_revenue():
    total = 0
    for book in inventory:
        total += book["quantity"] * book["price"]
    return total

def print_report():
    print("\n--- Inventory Report ---")
    for book in inventory:
        print("Title:", book["title"], 
              "Author:", book["author"], 
              "Quantity:", book["quantity"], 
              "Price:", book["price"])
    print("------------------------")

def check_stock():
    for book in inventory:
        if book["quantity"] == 0:
            print("ALERT:", book["title"], "is OUT OF STOCK!")

# Example
add_book("Python Basics", "John Smith", 5, 500)
add_book("Data Science 101", "Alice Brown", 0, 750)

print_report()
update_quantity("Python Basics", 10)
print("Total Revenue:", calculate_revenue())
check_stock()

```

students = []

# Input 3 students
for i in range(3):
    name = input("Enter name of student " + str(i+1) + ": ")
    scores = []
    for j in range(3):
        score = int(input("Enter score " + str(j+1) + " for " + name + ": "))
        scores.append(score)
    students.append([name, scores])

# Print scores
print("\n--- Student Scores ---")
for student in students:
    print(student[0], ":", student[1])

# Calculate averages
averages = []
for student in students:
    total = 0
    for score in student[1]:
        total += score
    avg = total / len(student[1])
    averages.append([student[0], avg])
    print("Average score of", student[0], "=", avg)

# Highest and lowest average
highest = averages[0]
lowest = averages[0]
for avg in averages:
    if avg[1] > highest[1]:
        highest = avg
    if avg[1] < lowest[1]:
        lowest = avg

print("\nHighest Average Score:", highest[0], "=", highest[1])
print("Lowest Average Score:", lowest[0], "=", lowest[1])

# Update score
name = input("\nEnter student name to update score: ")
found = False
for student in students:
    if student[0] == name:
        subject = int(input("Enter subject number (1-3): ")) - 1
        new_score = int(input("Enter new score: "))
        student[1][subject] = new_score
        print("Updated scores for", student[0], ":", student[1])
        found = True
        break
if not found:
    print("Student not found.")

# Sort by average (bubble sort style)
print("\n--- Sort by Average Score ---")
for i in range(len(averages)):
    for j in range(i+1, len(averages)):
        if averages[i][1] > averages[j][1]:
            averages[i], averages[j] = averages[j], averages[i]
for avg in averages:
    print(avg[0], "Average =", avg[1])

# Sort by name (bubble sort style)

names_sorted = students[:]
for i in range(len(names_sorted)):
    for j in range(i+1, len(names_sorted)):
        if names_sorted[i][0] > names_sorted[j][0]:
            names_sorted[i], names_sorted[j] = names_sorted[j], names_sorted[i]
for student in names_sorted:
    print(student[0], ":", student[1])

def division():

    try:
        num1 = float(input("Enter the first number to be divided"))
        num2 = float(input("Enter the second number to be divided"))

        answer = num1/num2
        print("The final value is: ",answer)

    except ValueError:
        print("The value is not a number!")

    except ZeroDivisionError:
        print("The second number cant be zero!")

    except TypeError:
        print("The entered values are not compatible for division!")

    except Exception as e:
        print("Unexpected error!")

    finally:
        print("Program finished...")

division()
