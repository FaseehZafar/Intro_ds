```python
#Task 3
import tkinter as tk

from tkinter import messagebox


questions = [
    "What is the capital of Pakistan?",
    "Which is the largest ocean?",
    "What is 5 + 7?"
]

options = [
    ["Karachi", "Lahore", "Islamabad", "Quetta"],
    ["Atlantic", "Indian", "Pacific", "Arctic"],
    ["10", "11", "12", "13"]
]

answers = [2, 2, 2]  

# App setup
root = tk.Tk()
root.title("Quiz App")
root.geometry("400x300")

score = 0
q_index = 0
selected_option = tk.IntVar()

def load_question():
    global q_index
    selected_option.set(-1)  
    question_label.config(text=questions[q_index])
    for i in range(4):
        radio_buttons[i].config(text=options[q_index][i])

def next_question():
    global q_index, score
    if selected_option.get() == answers[q_index]:
        score += 1
    q_index += 1
    if q_index < len(questions):
        load_question()
    else:
        messagebox.showinfo("Result", f"Your Score: {score}/{len(questions)}")
        root.destroy()

# Widgets
question_label = tk.Label(root, text="", font=("Arial", 14), wraplength=300, justify="center")
question_label.pack(pady=20)

radio_buttons = []
for i in range(4):
    rb = tk.Radiobutton(root, text="", variable=selected_option, value=i, font=("Arial", 12))
    rb.pack(anchor="w", padx=60)
    radio_buttons.append(rb)

next_btn = tk.Button(root, text="Next", command=next_question, font=("Arial", 12), bg="lightgreen")
next_btn.pack(pady=20)

# Start quiz
load_question()
root.mainloop()
```

```python
#Task 4
import csv
import tkinter as tk

# Step 1: Create CSV file with employee records
with open("employees.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Department", "Basic_Salary", "Bonus"])
    writer.writerow(["Ali", "IT", 50000, 5000])
    writer.writerow(["Sara", "HR", 40000, 3000])
    writer.writerow(["Ahmed", "Finance", 60000, 8000])
    writer.writerow(["Ayesha", "IT", 55000, 7000])

# Step 2: Read data and calculate Net Salary
employees = []
with open("employees.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        name = row["Name"]
        dept = row["Department"]
        basic = int(row["Basic_Salary"])
        bonus = int(row["Bonus"])
        net = basic + bonus
        employees.append({"Name": name, "Department": dept,
                          "Basic_Salary": basic, "Bonus": bonus,
                          "Net_Salary": net})

# Step 3: Find highest salary
highest = max(employees, key=lambda x: x["Net_Salary"])

# Step 4: Department-wise average salary
dept_totals = {}
dept_counts = {}
for emp in employees:
    d = emp["Department"]
    dept_totals[d] = dept_totals.get(d, 0) + emp["Net_Salary"]
    dept_counts[d] = dept_counts.get(d, 0) + 1

dept_avg = {d: dept_totals[d] / dept_counts[d] for d in dept_totals}

# Step 5: Save results into new CSV
with open("salary_results.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Department", "Basic_Salary", "Bonus", "Net_Salary"])
    for emp in employees:
        writer.writerow([emp["Name"], emp["Department"],
                         emp["Basic_Salary"], emp["Bonus"], emp["Net_Salary"]])
    writer.writerow([])
    writer.writerow(["Highest Salary", highest["Name"], highest["Net_Salary"]])
    writer.writerow([])
    writer.writerow(["Department", "Average Salary"])
    for d, avg in dept_avg.items():
        writer.writerow([d, avg])

print("✅ Salary results saved in 'salary_results.csv'")

# Step 6: Plot bar chart with Tkinter Canvas
root = tk.Tk()
root.title("Employee Salaries")

canvas = tk.Canvas(root, width=500, height=400, bg="white")
canvas.pack()

x = 50
bar_width = 50
max_salary = max(emp["Net_Salary"] for emp in employees)

for emp in employees:
    height = (emp["Net_Salary"] / max_salary) * 300
    canvas.create_rectangle(x, 350 - height, x + bar_width, 350, fill="blue")
    canvas.create_text(x + bar_width / 2, 360, text=emp["Name"])
    canvas.create_text(x + bar_width / 2, 350 - height - 10, text=str(emp["Net_Salary"]))
    x += 100

root.mainloop()
```
