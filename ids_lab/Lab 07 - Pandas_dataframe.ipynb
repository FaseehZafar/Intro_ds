### Data Transformation

```python
df = pd.DataFrame({
    'Name': ['Ali', 'Hamza', 'Talha'],
    'Age': [23, 23, 25],
    'Marks': [88, 92, 95]
})
df
```

```python
df['Marks'].apply(np.sqrt)
```

```python
df['Pass/Fail'] = pd.cut(df['Marks'], bins=[0,90,100], labels=['Fail','Pass'])
df
```

#### Grouping & Aggregation

```python
df.groupby('Age')['Marks'].mean()
```

```python
df.groupby('Age').agg({'Marks':['mean','max']})
```

```python
df = pd.DataFrame({
    'group': ['A', 'A', 'B', 'B', 'C'],
    'value': [10, 20, 30, 40, 50]
})

print("Original DataFrame:")
print(df)

# Use transform to calculate the mean per group
df['group_mean'] = df.groupby('group')['value'].transform('mean')

print("\nAfter transform:")
print(df)
```

### Merging, Joining, Concatenation

```python
df1 = pd.DataFrame({'ID':[1,2],'Name':['Ali','Hamza']})
df2 = pd.DataFrame({'ID':[1,2],'Marks':[88,92]})
merged = pd.merge(df1, df2, on='ID')
print(merged)
```

```python
pd.concat([df1, df2], axis=0)
```

```python
df1.join(df2, lsuffix='_left', rsuffix='_right')
```

### Pivot Tables & Crosstab

```python
pivot = pd.pivot_table(df, values='Marks', index='Age', columns='Name', aggfunc='mean')
pivot
```

```python
pd.crosstab(df['Age'], df['Marks'])
```

### Time Series

```python
dates = pd.date_range('2025-10-01', periods=5)
df_time = pd.DataFrame({'Date': dates, 'Value':[10,20,30,40,50]})
df_time.set_index('Date', inplace=True)
print(df_time.resample('D').mean())
```
